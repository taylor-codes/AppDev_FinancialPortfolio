```python
import quandl
import pandas as pd


# Not necessary, I just do this so I do not show my API key.
api_key = open('quandlapikey.txt','r').read()



df = quandl.get('NASDAQOMX/NQCIGCER', authtoken=api_key)
#         df.rename(columns={'Value':abbv}, inplace=True)

df = df.reset_index()
df = df[['Trade Date','Index Value']]   #this is crazy
```

```python
df = quandl.get('NASDAQOMX/NQCISIER', authtoken=api_key)#this is silver
df = df.reset_index()
df = df[['Trade Date','Index Value']]   #this is crazy
```

```python
df.to_pickle('RawSilver.pickle')
```

```python
df1=pd.read_pickle('RawGold.pickle')
df2=pd.read_pickle('RawSilver.pickle')
```

```python
print(df1.head())
print(df2.head())
```

```python
df = pd.merge(df1,df2, on='Trade Date')
```

```python
print(df.head())
```

```python
df.rename(columns={'Index Value_x':'Gold','Index Value_y':'Silver'},inplace=True)
```

```python
df.set_index('Trade Date',inplace=True)
```

```python
print(df.head())
```

```python
df.to_pickle('HeavyMetal.pickle')
```
